```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('j.jpg', 0)
_, img = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
cv.imwrite('th_img.jpg', img)
plt.imshow(img, 'gray')
res = img.copy()
```

```python
height, width = img.shape
for i in range(height):
    contours = []
    for j in range(width-1):
        if not img[i][j] == img[i][j+1]:
            contours.append([j+1, i])
    if contours:
        cv.circle(res, contours[0], 1, (100,100,100), -1, cv.LINE_AA)
        cv.circle(res, contours[-1], 1, (100,100,100), -1, cv.LINE_AA)
        
for j in range(width):
    contours = []
    for i in range(height-1):
        if not img[i][j] == img[i+1][j]:
            contours.append([j, i+1])
    if contours:
        cv.circle(res, contours[0], 1, (100,100,100), -1, cv.LINE_AA)
        cv.circle(res, contours[-1], 1, (100,100,100), -1, cv.LINE_AA)
        
plt.imshow(res, 'gray')
cv.imwrite('res.jpg', res)
```
