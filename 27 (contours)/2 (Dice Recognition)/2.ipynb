```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
org_img = cv.imread('2.jpg', 0)
img = org_img.copy()

_, img = cv.threshold(img, 230, 255, cv.THRESH_BINARY_INV)
img = cv.morphologyEx(img, cv.MORPH_CLOSE, np.ones((5,4)))

plt.imshow(img, 'gray')
cv.imwrite('22.jpg', img)
```

```python
contours, h = cv.findContours(img, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
hierarchy = h[0]
```

```python
parents=[hie[3] for hie in hierarchy]
for i in range(len(parents)):
    if parents[i] != -1:
        x, y, w, h = cv.boundingRect(contours[i])
        cv.rectangle(org_img, (x,y), (x+w,h+y), (0,0,0), 2)
plt.imshow(org_img, 'gray')
cv.imwrite('w.jpg', org_img)

dices=[]
for p in parents:
    if p != -1 and p not in dices:
        dices.append(p)
        
for dice in dices:
    print(parents.count(dice))
```
