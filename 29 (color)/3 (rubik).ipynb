```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('rubix.png')

def colors(pixel):
    count = 0
    if pixel[0] > 140: count+=1
    if pixel[1] > 140: count+=1
    if pixel[2] > 140: count+=1
    return count

for row in img:
    for pixel in row:
        if colors(pixel) == 2:
            for c in range(3):
                if pixel[c] > 140: pixel[c] = 0
                else: pixel[c] = 255
                
img=cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```
