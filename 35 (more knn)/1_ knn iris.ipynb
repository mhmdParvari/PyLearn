```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
x, y = sklearn.datasets.load_iris(return_X_y=True)
x, y = sklearn.utils.shuffle(x, y)
```

```python
cut = round(len(x) * .8)
x_train = x[ :cut]
y_train = y[ :cut]
x_test = x[cut: ]
y_test = y[cut: ]
```

```python
from my_knn import KNearestNeighbors
```

```python
accs = []
for k in [1 , 3 , 5 , 7 , 9 , 11 , 13 , 15]:
    knn = KNearestNeighbors(k)
    knn.fit(x_train, y_train)
    accs.append(knn.evaluate(x_test, y_test))
accs
```

```python
plt.plot(np.array([1, 3, 5, 7, 9, 11, 13, 15], str), accs, marker='o', markersize=3)
plt.xlabel('K')
plt.ylabel('accuracy')
plt.savefig('k.png')
```
