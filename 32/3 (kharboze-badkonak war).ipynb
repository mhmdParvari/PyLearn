```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 200

kharbozes = pd.DataFrame({
    'length': np.random.normal(40, 10, N),
    'width': np.random.normal(15, 5, N),
    'weight': np.random.normal(3000, 1000, N)}
)

badkonaks = pd.DataFrame({
    'length': np.random.normal(25, 7, N),
    'width': np.random.normal(22, 5, N),
    'weight': np.random.normal(50, 10, N)}
)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(kharbozes['weight'], kharbozes['length'], kharbozes['width'])
ax.scatter(badkonaks['weight'], badkonaks['length'], badkonaks['width'])

ax.set_xlabel('weight(gram)')
ax.set_ylabel('length(cm)')
ax.set_zlabel('width(cm)')
plt.show()
```
