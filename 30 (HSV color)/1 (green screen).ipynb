```python
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('SuperMan.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img = cv.cvtColor(img, cv.COLOR_RGB2HSV)
bkg = cv.imread('sky.jpg')
bkg = cv.cvtColor(bkg, cv.COLOR_BGR2RGB)
plt.imshow(bkg)
```

```python
height, width, _ = img.shape
bkg = cv.resize(bkg, (width, height))

h, s, v = cv.split(img)
img = cv.cvtColor(img, cv.COLOR_HSV2RGB)

for i in range(height):
    for j in range(width):
        if 38 < h[i][j] < 75 and s[i][j] > 98:
            img[i][j] = bkg[i][j]

plt.imshow(img)
```
