```python
import matplotlib.pyplot as plt
```

# Plot

```python
xpoints = [1,2,4]
ypoints = [2,3,1]
plt.plot(xpoints, ypoints, 'o--g', ms=10, lw=3)

plt.plot([1,2], [1,2], 'o')
```

# Subplot

```python
y = [2,1,0]
plt.subplot(1,2,1)
plt.plot(y)

y = [0,1,2]
plt.subplot(1,2,2)
plt.plot(y)
```

# Scatter

```python
x = [4,2,6,4]
y = [2,1,3,1]
plt.scatter(x,y, color='black', s=100)
```

# Bar

```python
x = ['737', 'a320', 'md82']
y = [5, 5, 8]

plt.bar(x, y , width=.3)
```

# Hist

```python
x = [1,2,3,4,1,3,1]
plt.hist(x)
plt.show()
```

# Pie

```python
vals = [10,20,30]
label = ['a','b','c']

plt.pie(vals, labels= label, explode=[.1,0,0], startangle= 30)
plt.show()
```
